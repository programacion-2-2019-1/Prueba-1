```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# t = np.arange(0.0, 2.0, 0.01)
# y1 = 1 + np.sin(2 * np.pi * t*3)
# y2 = .6 + np.cos(1 * np.pi * t*3)
dt = 0.05
t = np.arange(-2.0, 4.0, dt)
y3= t**3
# y3= 1 + np.sin(2 * np.pi * t)
y3= 1 + np.sin(2 * np.pi * t)*np.exp(-t)
cte = 1.1
y4= (y3>cte)*cte*2;  #print('t=',t,'\ty4=',y4)
y5= np.zeros(len(t))+cte; #print(y5)
fig, ax = plt.subplots()
ax.plot(t,y3,t,y4,t,y5)
ax.set(xlabel='Eje X', ylabel='Eje Y',title='Ejemplo')
ax.grid()
fig.savefig("test.png")
plt.show()
```
