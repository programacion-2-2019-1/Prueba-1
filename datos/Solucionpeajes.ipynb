```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Como primer paso, leámos el data frame y hagamosle una pequeña "limpieza":

```python
df= pd.read_csv("Tabla_de_Costos_de_Peajes.csv")
df=df.fillna(0)
```

Calculemos entonces el número de peajes por departamento:

```python
dfDep=df['Departamento']
x2=dfDep.unique()
y=np.zeros(len(x2))
for i in range(len(x2)):
    reg = x2[i]
    y[i]=len(df.query('Departamento==@reg'))
    print('El número de peajes de',reg.lower(),'es',int(y[i]))
p1=plt.barh(x2,y,color='r')
```

Ahora, veamos cuántos proyectos hay por departamento:

```python
dfProy=df[['NombreProyecto','Departamento']]
dfProy= dfProy.drop_duplicates()
y1=np.zeros(len(x2))
for j in range(len(x2)):
    NomProy = x2[j]
    y1[j]=len(dfProy.query('Departamento==@NomProy'))
    print('El número de proyectos en',NomProy.lower(),'es',int(y1[j]))
p2=plt.barh(x2,y1)
```

Ahora veamos cuánto hace cada departamento con los peajes

```python
dfPeajes=df.loc[:,'Departamento':'TAR_ESPECIAL_E.C.']
dfPeajes=dfPeajes.drop(['Municipio'], axis=1)
y2=np.zeros(len(x2))
for j1 in range(len(x2)):
    reg=x2[j1]
    dftemp=dfPeajes.loc[dfPeajes['Departamento'] == reg]
    dftemp=dftemp.drop(['Departamento'],axis=1)
    idx=list(dftemp.index) 
    con=np.zeros(len(idx))
    for j2 in range(len(idx)):
        tp=dftemp.loc[idx[j2],:]
        con[j2]=sum(tp)
    print('La suma total de los costos de los peajes de ',reg.lower(),'es',int(sum(con))) 
    y2[j1]=int(sum(con))
y2=[i/1000 for i in y2]
p3=plt.barh(x2,y2,color='k')
```

Por último, veamos los costes de peajes más caros:

```python
dfCost=df['TAR_PLENA_VII'];
dfCost=dfCost.sort_values(ascending=False)
dfCost=dfCost.unique()
dfCost.sort()
y3=dfCost[::-1]
for jf in range(0,9):
    print(y3[jf])
xf=range(1,10)
p4=plt.bar(xf,y3[0:9])
```
