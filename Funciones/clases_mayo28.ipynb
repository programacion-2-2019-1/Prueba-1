```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from clase2 import *
print('clases P-II')
```

Se crea figura1 y figura2 como **instancias** (objetos) de la clase Figura. 

```python
figura1 = Figura()
figura2 = Figura()
print('Clase creada')
```

```python
figura1.rotar(3.1415/2)
figura2.rotar(-3.1415/4)
dt = 0.5
t = np.arange(-2*np.pi,2*np.pi,dt)
y = 2*np.sin(t)
```

```python
fig, ax=plt.subplots()
ax.plot(t,y,'go-')
ax.plot(figura1.x,figura1.y,'ko-')
ax.plot(figura2.x,figura2.y,'ro-')
figura1.rotar(3.1415/2)
figura1.reset()
ax.plot(figura1.x,figura1.y,'bo-')

ax.grid(); ax.axis([-5,5,-3,3])
plt.show
print('Fin ok')
```
